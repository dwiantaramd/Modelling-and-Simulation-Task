```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython import display 

#Inisiasi Variabel Konstanta / Input
x0 = 0      # Posisi x awal
y0 = 0      # Posisi y awal
v0 = 50     # Kecepatan awal
t0 = 0      # waktu
g = -9.806  # Gravitasi
m = 0.15    # Massa
D = 0.0013  # Konstanta
sudut = 35  # sudut
tstep = 0.01   # delta t / time step

radian = (sudut/360)*(2*np.pi) # sudut dalam radian
```

# Solusi Numerik Tanpa Hambatan Udara

```python
# Tanpa hambatan
vx = v0 * np.cos(radian)       # kecepatan x awal
vy = v0 * np.sin(radian)       # kecepatan y awal

# Inisiasi array posisi dan waktu
x_numerik = [x0]
y_numerik = [y0]
t_numerik = [t0]

# Solusi Numerik
i = 0
while y_numerik[i] >= 0:
    vy = vy + g*tstep
    y_next = y_numerik[i] + vy*tstep
    x_next = x_numerik[i] + vx*tstep
    t_next = t_numerik[i] + tstep
    x_numerik.append(x_next)
    y_numerik.append(y_next)
    t_numerik.append(t_next)
    i = i + 1
```

# Solusi Numerik dengan Hambatan Udara

```python
# Dengan hambatan udara
vx_hambatan = v0 * np.cos(radian) # kecepatan x awal
vy_hambatan = v0 * np.sin(radian) # kecepatan y awal


# Inisiasi array posisi dan waktu
x_hambatan = [x0]
y_hambatan = [y0]
t_hambatan = [t0]

# Solusi Numerik
i = 0
while y_hambatan[i] >= 0:
    v = math.sqrt(vx_hambatan**2 + vy_hambatan**2)
    ax_hambatan = -1 * (D/m) * v * vx_hambatan
    ay_hambatan = g - (D/m) * v * vy_hambatan
    vx_hambatan = vx_hambatan + (ax_hambatan * tstep)
    x_next = x_hambatan[i] + (vx_hambatan * tstep)
    vy_hambatan = vy_hambatan + (ay_hambatan * tstep)
    y_next = y_hambatan[i] + (vy_hambatan * tstep)
    t_next = t_hambatan[i] + tstep
    
    x_hambatan.append(x_next)
    y_hambatan.append(y_next)
    t_hambatan.append(t_next)
    i = i + 1
  
```

# Solusi Analitik Tanpa Hambatan Udara

```python
# Inisiasi array posisi dan waktu
x_analitik = [x0]
y_analitik = [y0]
t_analitik = [t0]

# Solusi Analitik
i = 0
while y_analitik[i] >= 0:
    x_next = v0 * np.cos(radian) * t_analitik[i]
    y_next = (0.5 * g * t_analitik[i]**2) + (v0 * np.sin(radian) * t_analitik[i])
    x_analitik.append(x_next)
    y_analitik.append(y_next)
    t_analitik.append(t_analitik[i] + tstep)
    i = i + 1
```
