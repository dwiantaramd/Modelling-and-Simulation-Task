```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
total = 100 # Jumlah Dart
iteration = 100 #Jumlah Iterasi
pi_arr = []
x_hit = []
y_hit = []
x_not = []
y_not = []
```

```python
def monteCarlo(total, x_hit, y_hit, x_not, y_not):
    hit = 0
    ndart = 0
    while ndart < total:
        Xrand = random.uniform(0,1)
        Yrand = random.uniform(0,1)
        ndart += 1
        # Check Condition
        if (Xrand**2)+(Yrand**2)<=1:
            hit += 1
            x_hit.append(Xrand)
            y_hit.append(Yrand)
        else:
            x_not.append(Xrand)
            y_not.append(Yrand)
            
    return hit, x_hit, y_hit, x_not, y_not
```

```python
for i in range (0, iteration):
    hit, x_hit, y_hit, x_not, y_not = monteCarlo(total, x_hit, y_hit, x_not, y_not)
    pi = (hit/total)*4
    pi_arr.append(pi)
```

```python
theta = np.linspace(0, 2*np.pi, 100)
r = np.sqrt(1.0)
x1 = r*np.cos(theta)
x2 = r*np.sin(theta)
fig, ax = plt.subplots(1)
ax.plot(x1, x2)
ax.scatter(x_hit, y_hit, s=0.4)
ax.scatter(x_not, y_not, s=0.4)
ax.set_aspect(1)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title('N=10000')
plt.show()
```

```python
plt.figure()
plt.plot(pi_arr, label='pi Estimasi')
plt.ylim(0, 5)
plt.plot([0,100], [math.pi, math.pi], label='pi Eksak')
plt.legend()
plt.show()
```
